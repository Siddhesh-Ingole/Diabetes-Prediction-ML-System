```python
# Import Libraries
import pandas as pd
import joblib
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix


# CONTROL FLAG

SHOW_EVALUATION = False


# Dataset

data_dict = {
"Pregnancies":[6,1,8,1,0,5,3,10,2,8],
"Glucose":[148,85,183,89,137,116,78,115,197,125],
"BloodPressure":[72,66,64,66,40,74,50,0,70,96],
"SkinThickness":[35,29,0,23,35,0,32,0,45,0],
"Insulin":[0,0,0,94,168,0,88,0,543,0],
"BMI":[33.6,26.6,23.3,28.1,43.1,25.6,31,35.3,30.5,0],
"DiabetesPedigreeFunction":[0.627,0.351,0.672,0.167,2.288,0.201,0.248,0.134,0.158,0.232],
"Age":[50,31,32,21,33,30,26,29,53,54],
"Outcome":[1,0,1,0,1,0,0,1,1,1]
}

data = pd.DataFrame(data_dict)


# Split Data

X = data.drop("Outcome", axis=1)
Y = data["Outcome"]

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)

# Feature Scaling

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)


# Train Model

model = LogisticRegression(max_iter=1000)
model.fit(X_train, Y_train)

# Evaluation (Hidden by default)

y_pred = model.predict(X_test)

if SHOW_EVALUATION:
    print("\nModel Evaluation")
    print("Accuracy:", accuracy_score(Y_test, y_pred))
    print("Confusion Matrix:\n", confusion_matrix(Y_test, y_pred))


# Save Model

joblib.dump(model, "diabetes_model.pkl")


# History

history = []


# Main Loop

while True:

    print("\n1. Get Prediction")
    print("2. View Prediction History")
    print("3. Save History to CSV")
    print("4. Exit")

    try:
        option = int(input("Enter Option: "))
    except:
        print("Invalid input")
        continue

    if option == 1:

        print("\nEnter Personal Details")
        name = input("Name: ")
        contact = input("Contact: ")
        gender = input("Gender (Male/Female): ").lower()

        print("\nEnter Medical Details")

        try:
            # Gender-based Pregnancies Fix
            if gender == "female":
                pregnancies = float(input("Pregnancies: "))
            elif gender == "male":
                pregnancies = 0
                print("Pregnancies set to 0 for male")
            else:
                print("Invalid gender")
                continue

            glucose = float(input("Glucose: "))
            bp = float(input("BloodPressure: "))
            skin = float(input("SkinThickness: "))
            insulin = float(input("Insulin: "))
            bmi = float(input("BMI: "))
            dpf = float(input("DiabetesPedigreeFunction: "))
            age = float(input("Age: "))

        except:
            print("Invalid medical input")
            continue

        input_data = [[pregnancies, glucose, bp, skin, insulin, bmi, dpf, age]]

        # Scaling
        input_data = scaler.transform(input_data)

        prediction = model.predict(input_data)
        probability = model.predict_proba(input_data)

        diabetic_prob = probability[0][1]
        non_diabetic_prob = probability[0][0]

        result = "Diabetic" if prediction[0] == 1 else "Non-Diabetic"

        print("\n Prediction Result")
        print("Result:", result)
        print("Probability Diabetic:", round(diabetic_prob, 2))
        print("Probability Non-Diabetic:", round(non_diabetic_prob, 2))


        # Correct Note Logic

        if result == "Diabetic":
            print("\n Note: You are diabetic. Please consult a doctor for further treatment.")
        else:
            print("\n Note: You are non-diabetic. You are healthy. Keep maintaining a healthy lifestyle.")

        # Save to history
        history.append({
            "Name": name,
            "Contact": contact,
            "Gender": gender,
            "Result": result
        })

    elif option == 2:

        print("\n Prediction History")

        if len(history) == 0:
            print("No history yet")
            continue

        for i, h in enumerate(history, 1):
            print(i, h)

    elif option == 3:

        pd.DataFrame(history).to_csv("history.csv", index=False)
        print(" History saved to history.csv")

    elif option == 4:
        print("Program Closed")
        break

    else:
        print(" Invalid Option")
```
